# Data Analysis & Visualizations

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)
plt.figure(figsize = (5,4))
plt.title('reservation status count')
plt.bar(['not canceled','canceled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
plt.show()
```

```python
plt.figure(figsize = (10,6))
ax1 = sns.countplot(x = 'hotel', hue ='is_canceled',data = df,palette='Blues')
legend_labels,_ = ax1.get_legend_handles_labels()
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average daily Rate In City And Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
df['country'].value_counts()[:10]
```

```python
plt.figure(figsize=(8,6), dpi=110)
sns.countplot(data=df, x='country',order=pd.value_counts(df['country']).iloc[:10].index, palette='colorblind')
plt.title('Top 10 Countries of Origin of the Guests', weight='bold')
plt.xlabel('Country')
plt.ylabel('Reservation Count')
```

This graph show information about the countries that have most bookings. Most of bookings are created from Portugal and then Great Britain in second place and France.

```python
month_sorted = ['January','February','March','April','May','June','July','August','September','October','November','December']
plt.figure(figsize=(14,6))
sns.countplot(data=df ,x='arrival_date_month',hue = 'is_canceled', palette='pastel', order=pd.value_counts(df['arrival_date_month']).index)
plt.title('Busiest Month with Bookings')
plt.xlabel('Month')
plt.ylabel('Reservation Count')
plt.legend(['not canceled','canceled'])
plt.show()
```

The above graph depicts the number of bookings created for each month
According to the graph, August is the most occupied (busiest) month
